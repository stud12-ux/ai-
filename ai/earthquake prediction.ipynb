```python
df['date_time'] = pd.to_datetime(df['date_time'], format='%d-%m-%Y %H:%M')

df['date_time'] = pd.to_datetime(df['date_time'], format='%d/%m/%Y %H:%M')

df['year'] = df['date_time'].dt.year
df['month'] = df['date_time'].dt.month
df['day'] = df['date_time'].dt.day
df['hour'] = df['date_time'].dt.hour
```

```python
df = df.drop(columns=['date_time'])


```

```python
df = df.drop(columns = ["title"])
```

```python
x = df.drop(columns = ["magnitude"])

y =df['magnitude'].astype(float) 
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

```python
x_train.head(4)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
```

```python
category_column = [ 'magType', 'net', 'location', 'continent', 'country']

numeric_columns = ['cdi', 'mmi', 'tsunami', 'sig', 'nst', 'dmin', 'gap', 'depth', 'latitude', 'longitude', 'year', 'month', 'day', 'hour']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), category_column),
        ('num', StandardScaler(), numeric_columns)
    ])
```

```python
model = Pipeline(steps=[
   ('preprocessor' , preprocessor),
    ('regressor' , LinearRegression()),
    #('classifier', SVC(kernel='linear'))
])
```

```python
numeric_columns
```

```python
model.fit(x_train, y_train)
```
